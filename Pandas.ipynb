```python
test_birthdate = pd.to_datetime('1992-01-31')
test_today = pd.to_datetime('2024-01-15')

test_birthdate_year = test_birthdate.year
test_today_year = test_today.year

year_difference = test_today_year - test_birthdate_year
year_difference
```

```python
month_day = (test_today.month, test_today.day) < (test_birthdate.month, test_birthdate.day)
age = year_difference - month_day
age
```

```python
def calculate_age(birthdate):
    todays_date = pd.to_datetime('today')

    birthdate_year = birthdate.year
    today_year = todays_date.year

    year_diff = today_year - birthdate_year

    age = year_diff - ((todays_date.month, todays_date.day) < (birthdate.month, birthdate.day))
    return age
```

```python
rewards_df_sample['Age'] = rewards_df_sample['Birthdate'].apply(calculate_age)
rewards_df_sample
```
